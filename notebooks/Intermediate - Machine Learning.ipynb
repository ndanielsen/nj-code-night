```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Read the data file ##

```python
# read the iris data into a pandas DataFrame, including column names
col_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                   names=col_names)

iris['species_num'] = iris.species.map({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
```

## Format the Data ##


```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
classes = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']


X = iris[features].as_matrix()
y = iris.species_num.as_matrix()
```

# Step 1: import the class you plan to use


Classifier implementing the k-nearest neighbors vote.

http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

```python
from sklearn.neighbors import KNeighborsClassifier
```

# Step 2: instantiate the "estimator" (aka the model)

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

# Step 3: fit the model with data (learn the relationship between X and y)

```python
knn.fit(X, y)
```

# Step 4: use the "fitted model" to predict the response for a new observation

```python
knn.predict([[3, 5, 4, 2]])

# predict for multiple observations at once
X_new = [[3, 5, 4, 2], [3, 5, 2, 2]]
knn.predict(X_new)
```

# Evaluate

```python
from yellowbrick.classifier import DecisionViz

features = ['sepal_length', 'sepal_width']
classes = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

X = iris[features].as_matrix()
y = iris.species_num.as_matrix()

knn = KNeighborsClassifier(n_neighbors=10)

viz = DecisionViz(knn, classes=classes, features=features)

viz.fit(X, y)
viz.draw(X, y)
viz.poof()
```

```python
from yellowbrick.classifier import DecisionViz

features = ['petal_length', 'petal_width']
classes = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

X = iris[features].as_matrix()
y = iris.species_num.as_matrix()

knn = KNeighborsClassifier(n_neighbors=5)

viz = DecisionViz(knn, classes=classes, features=features)

viz.fit(X, y)
viz.draw(X, y)
viz.poof()
```
